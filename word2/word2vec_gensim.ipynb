```python
#词向量
import pandas as pd
import jieba
from gensim.models.word2vec import Word2Vec

# 读入训练集文件
data = pd.read_csv('train.csv')
# 转字符串数组
corpus = data['comment'].values.astype(str)
# 分词，再重组为字符串数组
corpus = [jieba.lcut(corpus[index]
                          .replace("，", "")
                          .replace("!", "")
                          .replace("！", "")
                          .replace("。", "")
                          .replace("~", "")
                          .replace("；", "")
                          .replace("？", "")
                          .replace("?", "")
                          .replace("【", "")
                          .replace("】", "")
                          .replace("#", "")
                        ) for index in range(len(corpus))]
# 词向量模型训练
model = Word2Vec(corpus, sg=0, vector_size=300, window=5, min_count=3, workers=4)
#模型显示
print('模型参数：',model,'\n')
```

```python
# 输出“环境”的词向量
env_vector = model.wv['环境']
print("词向量：", env_vector)

# 输出词向量的形状
print("词向量形状：", env_vector.shape)
```

```python
# 输出与“好吃”语义最接近的3个词
similar_words = model.wv.most_similar('好吃', topn=3)
print("与'好吃'最接近的词：", similar_words)
```

```python
# 计算相似度
similarity_delicious_tasty = model.wv.similarity('好吃', '美味')
similarity_delicious_cockroach = model.wv.similarity('好吃', '蟑螂')

print("好吃和美味的相似度：", similarity_delicious_tasty)
print("好吃和蟑螂的相似度：", similarity_delicious_cockroach)
```

```python
# 执行向量运算
result = model.wv.most_similar(positive=['餐厅', '聚会'], negative=['安静'], topn=1)
print("餐厅+聚会-安静的结果：", result)
```
